Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Seaborn is a popular Python data visualization library built on top of Matplotlib. It provides a high-level interface for creating informative and attractive statistical graphics. Here are five common plots that you can create using Seaborn along with their typical uses:

Scatter Plot:

Uses: Scatter plots display the relationship between two continuous variables. They are useful for identifying patterns, clusters, or correlations in the data.

Line Plot:

Uses: Line plots are typically used to visualize the trend or evolution of a continuous variable over time. They are useful for analyzing time series data.

Bar Plot:

Uses: Bar plots are used to compare categories or groups. They represent categorical data using rectangular bars, where the length or height of each bar corresponds to the quantity being plotted.

Histogram:

Uses: Histograms display the distribution of a continuous variable. They group data into bins and represent the frequency or count of observations within each bin.

Heatmap:

Uses: Heatmaps are used to visualize the magnitude of a quantitative variable in a matrix or grid-like format. They are particularly useful for displaying correlations or relationships between variables.

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.
Note: timepoint, signal, event, and region are columns in the fmri dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "fmri" dataset
fmri = sns.load_dataset("fmri")

# Create a line plot using Seaborn
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event", style="region")


plt.show()
```

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "titanic" dataset
titanic = sns.load_dataset("titanic")

# Create the first box plot for 'age'
plt.figure(figsize=(10, 6))  
sns.boxplot(data=titanic, x='pclass', y='age')
plt.title("Box Plot of Age by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Age")

# Create the second box plot for 'fare'
plt.figure(figsize=(10, 6)) 
sns.boxplot(data=titanic, x='pclass', y='fare')
plt.title("Box Plot of Fare by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Fare")

# Display the plots
plt.show()
```

Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

diamonds = sns.load_dataset("diamonds")
diamonds.head()

```

```python

plt.figure(figsize=(10, 6))  
sns.histplot(data=diamonds, x="price", hue="cut", kde=True)
plt.title("Histogram of Diamond Prices by Cut")
plt.xlabel("Price")
plt.ylabel("Count")

plt.show()
```

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "flights" dataset
flights = sns.load_dataset("flights")
flights.head()
```

```python
flights_pivot = flights.pivot("month", "year", "passengers")

plt.figure(figsize=(10, 6)) 
sns.heatmap(data=flights_pivot, cmap="YlGnBu", annot=True, fmt="d")
plt.title("Passenger Count by Month and Year")
plt.xlabel("Year")
plt.ylabel("Month")

plt.show()
```
